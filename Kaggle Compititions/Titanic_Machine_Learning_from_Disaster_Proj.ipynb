##Kaggle-Titanic - Machine Learning from Disaster

```python
#Importing required
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
#EDA-Ecploratory Data Analysis
#Load the Dataset
titanic_train_df=pd.read_csv('/content/train.csv')
titanic_test_df=pd.read_csv('/content/test.csv')
print("Shape of titanic_train_df",titanic_train_df.shape)
print("Shape of titanic_test_df",titanic_test_df.shape)
```

```python
titanic_train_df.head()
```

```python
titanic_train_df.isnull().sum()
```

```python
titanic_test_df.isnull().sum()
```

```python
titanic_train_df['Age'].fillna(titanic_train_df['Age'].mean(),inplace=True)
```

```python
titanic_test_df['Age'].fillna(titanic_test_df['Age'].mean(),inplace=True)
titanic_test_df['Fare'].fillna(titanic_test_df['Fare'].mean(),inplace=True)
```

```python
print("Shape of titanic_train_df",titanic_train_df.shape)
print("Shape of titanic_test_df",titanic_test_df.shape)
```

```python
titanic_train_df.isnull().sum()
titanic_test_df.isnull().sum()
```

```python
titanic_train_df.head()
```

```python
#prepare X_train,y_train,X_test,y_test
#Feature Engneering
#Converted 'Sex', 'Embarked', and 'Title' into numbers using one-hot encoding.
#Sex
titanic_train_df['Sex'] = titanic_train_df['Sex'].str.lower().str.strip().map({'male': 0, 'female': 1})
titanic_test_df['Sex'] = titanic_test_df['Sex'].str.lower().str.strip().map({'male': 0, 'female':1})
#Embarked
titanic_train_df['Embarked'] = titanic_train_df['Embarked'].str.lower().str.strip().map({'c': 0, 'q': 1,'s':2})
titanic_test_df['Embarked'] = titanic_test_df['Embarked'].str.lower().str.strip().map({'c': 0, 'q': 1,'s':2})
#handining nan values
titanic_train_df['Embarked'].fillna(titanic_train_df['Embarked'].mean(),inplace=True)
#Title

#Adding new column 'Family Size' by combining SibSp and Parch for both train and test df
titanic_train_df['Family_size']=titanic_train_df['SibSp']+titanic_train_df['Parch']
titanic_test_df['Family_size']=titanic_test_df['SibSp']+titanic_test_df['Parch']
```

```python
titanic_train_df.head()
```

```python
# prompt: Adding isALone column for finding is the pasenger traveling alone as 1 and if not alone then 0

# Adding 'isAlone' column
titanic_train_df['isAlone'] = 0
titanic_train_df.loc[titanic_train_df['Family_size'] == 0, 'isAlone'] = 1

titanic_test_df['isAlone'] = 0
titanic_test_df.loc[titanic_test_df['Family_size'] == 0, 'isAlone'] = 1
```

```python
titanic_train_df.head()
```

```python
print("Shape of titanic_train_df",titanic_train_df.shape)
print("Shape of titanic_test_df",titanic_test_df.shape)
```

```python
#Defining X_train,y_train
X_train=titanic_train_df[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Family_size','isAlone']]
y_train=titanic_train_df['Survived']
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
#Defining y_test
X_test=titanic_test_df[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Family_size','isAlone']]
X_test.shape
```

```python
# Model Preperation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
#train the model
model.fit(X_train,y_train)
#test the model
y_pred=model.predict(X_test)
y_pred
```

```python
y_pred_df=pd.DataFrame(y_pred)
y_pred_df = (y_pred_df >= 0.6).astype(int)
test_PassengerIds=titanic_test_df['PassengerId']
```

```python
# Ensure both are 1D Series
y_pred_df = y_pred_df.squeeze().reset_index(drop=True)
test_PassengerIds = test_PassengerIds.squeeze().reset_index(drop=True)

# Combine into a DataFrame
gender_submission_Final_pridiction = pd.DataFrame({
    'PassengerId': test_PassengerIds,
    'Survived': y_pred_df
})


gender_submission_Final_pridiction
```

```python

gender_submission_Final_pridiction.to_csv('gender_submission_Final_pridiction.csv', index=False)
```
